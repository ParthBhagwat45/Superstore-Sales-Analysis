```python
import pandas as pd

df = pd.read_csv("Superstore dataset.csv")
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.dtypes
```

```python
negative_sales = sum(df['Sales'] < 0)
negative_discount = sum(df['Discount'] < 0)
invalid_quantity = sum(df['Quantity'] <= 0)
negative_profit = sum(df['Profit'] < 0)

print(f"Negative Sales Records: {negative_sales}")
print(f"Negative Discount Records: {negative_discount}")
print(f"Invalid Quantity Records (<=0): {invalid_quantity}")
print(f"Negative Profit Records: {negative_profit}")
```

```python
print(df['Order Date'].min())
print(df['Order Date'].max())
```

```python
print("=== Categorical Data Validation ===")

print(f"\nRegions: {df['Region'].unique()}")

print(f"\nCustomer Segments: {df['Segment'].unique()}")

print(f"\nProduct Categories: {df['Category'].unique()}")

print(f"\nShipping Modes: {df['Ship Mode'].unique()}")
```

```python
df['Year'] = df['Order Date'].dt.year

df['Month'] = df['Order Date'].dt.month

df['Month Name'] = df['Order Date'].dt.month_name()

df['Quarter'] = df['Order Date'].dt.quarter

df['Day of Week'] = df['Order Date'].dt.day_name()

df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

df.head()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()
average_order_value = total_sales / total_orders
profit_margin = (total_profit / total_sales) * 100

print(f"Total Sales: {total_sales:.2f}")
print(f"Total Profit: {total_profit:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Total Customers: {total_customers}")
print(f"Average Order Value: {average_order_value:.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
sales_by_year = df['Sales'].groupby(df['Year']).sum()
print(sales_by_year)
```

```python
profit_by_year = df['Profit'].groupby(df['Year']).sum()
print(profit_by_year)
```

```python
profit_margin_by_year = (profit_by_year/sales_by_year)*100
print(f'Profit percent over the years:\n{profit_margin_by_year}')
```
